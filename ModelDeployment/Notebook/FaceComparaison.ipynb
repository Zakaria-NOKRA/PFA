```python
from deepface import DeepFace
import pickle
import os
```

```python
model_name = 'Facenet512'
model = DeepFace.build_model(model_name)
```

```python
folder_path = "./../SavedModel"
```

```python
model_path = os.path.join(folder_path,'deepface_facenet_model.pkl')
```

```python
with open(model_path,'wb') as f:
    pickle.dump(model,f)
```
